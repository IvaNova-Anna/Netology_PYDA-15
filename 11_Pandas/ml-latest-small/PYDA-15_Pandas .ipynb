### Задача 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.


```python
data = pd.read_csv('power.csv')
countries_filter = data[ (data['country']=='Latvia') | (data['country']=='Lithuania') | (data['country']=='Estonia')]
new_data = countries_filter.sort_values(by=['year', 'category' ], ascending=[True, True])

a = new_data[ (new_data.year >= 2005) & (new_data.year <= 2010) & 
            ((new_data.category == 4) | (new_data.category == 12) | (new_data.category == 21)) & 
            (new_data.quantity >= 0)]
sum_ = round (countries_filter.quantity.sum())
print('Table. Суммарное потребление товаров странами Прибалтики (Латвия, Литва и Эстония) за период с 2005 по 2010 года. ')
print('Страны Прибалтики потребляют суммарно =', sum_)
a.groupby(['year','category'])['quantity'].sum().reset_index()
```
